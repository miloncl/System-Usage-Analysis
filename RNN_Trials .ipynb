```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler

# load the data into a pandas dataframe
df = pd.read_csv('/content/drive/MyDrive/DSC180B/Dsc180b-main 2/chrome_time.csv')
df = df.drop(['Unnamed: 0'], axis = 1)

scaler = MinMaxScaler()
df['Total_Usage'] = scaler.fit_transform(df[['Total_Usage']])
# split the data into features (X) and target (y)
X = df.iloc[:, :24].values
y = df.Total_Usage.values

# reshape the features for the RNN
X = np.reshape(X, (X.shape[0], 1, X.shape[1]))

# build the RNN model
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(LSTM(16))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

# compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# fit the model to the training data
model.fit(X, y, epochs=100, batch_size=32, verbose=0)

# make predictions using the model
y_pred = model.predict(X)

# add the predictions as a column in the dataframe
df['PRED'] = y_pred.flatten()
df['PRED'] = scaler.fit_transform(df[['PRED']])

# calculate false positive, false negative, true positive and true negative
threshold = 0.5

df['TP'] = np.where((df['Total_Usage'] == 1) & (df['PRED'] > threshold), 1, 0)
df['TN'] = np.where((df['Total_Usage'] == 0) & (df['PRED'] < threshold), 1, 0)
df['FP'] = np.where((df['Total_Usage'] == 0) & (df['PRED'] > threshold), 1, 0)
df['FN'] = np.where((df['Total_Usage'] == 1) & (df['PRED'] < threshold), 1, 0)

TP = df['TP'].sum()
TN = df['TN'].sum()
FP = df['FP'].sum()
FN = df['FN'].sum()

# calculate accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)

# calculate false positive rate
false_positive_rate = FP / (FP + TN)

# calculate false negative rate
false_negative_rate = FN / (TP + FN)

# calculate true positive rate
true_positive_rate = TP / (TP + FN)

# calculate true negative rate
true_negative_rate = TN / (TN + FP)

print('Accuracy: ', accuracy)
print('False Positive Rate: ', false_positive_rate)
print('False Negative Rate: ', false_negative_rate)
print('True Positive Rate: ', true_positive_rate)
print('True Negative Rate: ', true_negative_rate)

df.head()
```
